```python
df['Cpu_brand'] = df['Cpu'].apply(lambda x:''.join(x.split()[:3]))

def fetch(text):
    if text=='InteL Core i3' or text== 'Intel Core i5' or text=='InteL Core i7':
        return text
    else:
        if text.split()[0]=='Intel':
            return 'Other intel'
        else:
            return 'AMD'

df['Cpu_brand'] = df['Cpu_brand'].apply(fetch)
```

```python
df['HDD'] = 0
df['SSD'] = 0
df['Flash'] = 0
df['Hybrid'] = 0
```

```python
import re

def extract_memory(mem):
    hdd = 0
    ssd = 0
    flash = 0
    hybrid = 0

    parts = mem.split(" + ")

    for part in parts:

        size = int(re.findall(r'\d+', part)[0])

        if "TB" in part:
            size *= 1024

        if "SSD" in part:
            ssd = size

        elif "HDD" in part:
            hdd = size

        elif "Flash Storage" in part:
            flash = size

        elif "Hybrid" in part:
            hybrid = size

    return pd.Series([hdd, ssd, flash, hybrid])
```

```python
df[['HDD','SSD','Flash','Hybrid']] = df['Memory'].apply(extract_memory)
```
