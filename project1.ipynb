```python
import numpy as np
import pandas as pd

df = pd.read_csv('laptop_data.csv')
```

```python
df.head()
```

```python
#removing unnamed column
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
#removing gb from ram and kg from weight and changing there dtype to int32 and float32
df['Ram'] = df['Ram'].str.replace("GB","")
df['Weight'] = df['Weight'].str.replace("kg","")

df['Ram'] = df['Ram'].astype('int32')
df['Weight'] = df['Weight'].astype('float32')
```

```python
#changing dtype of price to int
df['Price'] = df['Price'].astype('int32')
```

```python
df.info()
```

```python
df.head()
```

```python
new = df['ScreenResolution'].str.split('x',n=1,expand=True)
df['x_res'] = new[0]
df['y_res'] = new[1]
```

```python
df['x_res'] = df['x_res'].str.extract(r'(\d+)')
```

```python
df['x_res'] = df['x_res'].astype('int32')
df['y_res'] = df['y_res'].astype('int32')
```

```python
df['ppi'] = (((df['x_res']**2) + (df['y_res']**2))**0.5/df['Inches'])
```

```python
df.drop(columns=['Inches','x_res','y_res','ScreenResolution'],inplace=True)
```
